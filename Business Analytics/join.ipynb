```python
import pandas as pd
```

```python
#Load the data
rep = pd.read_csv("reps.csv")
dem14 = pd.read_csv("demographics2014_change.csv")
dem16 = pd.read_csv("demographics2016_change.csv")
dem18 = pd.read_csv("demographics2018_change.csv")
```

```python
#prepare the data to be merged
def prepare(df,year):
    df = df.rename(columns = {"state_code":"state_fips"})
    df["year"] = year
    return df
    
```

```python
#make the various tables uniform
dem14 = prepare(dem14,2014)
dem16 = prepare(dem16,2016)
dem18 = prepare(dem18,2018)

#Combine all the demographic data
appended = (dem14.append(dem16)).append(dem18)

#Join/merge the representative data with the demogrpahic data
merged = rep.merge(appended, on=["state_fips", "district","year"])
```

```python
merged.head()
```

```python
#drop some unneccesary columns
final = merged.drop(["state_po","state_cen", "state_ic", "office", "stage", "runoff", "special", "writein", "mode",
             "candidatevotes", "totalvotes", "unofficial", "version", "fusion_ticket"], axis=1)
```

```python
final.to_csv("final.csv", index=False)
```
